```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

camera = cv2.VideoCapture("new.avi")

# here are all the kinds of tracker

tracker = cv2.TrackerBoosting_create() 
#tracker = cv2.TrackerMIL_create()      
#tracker = cv2.TrackerKCF_create()  
#tracker = cv2.TrackerTLD_create()      
#tracker = cv2.TrackerMedianFlow_create() 
#tracker = cv2.TrackerMOSSE_create()    
#tracker = cv2.TrackerCSRT_create()      


ret,image = camera.read()
image = cv2.resize(image,(500,500))
bbox = cv2.selectROI(image ,False)

tracker.init(image,bbox)

#get the coordiates and the image and draw rectangle on it
def drawBox(image,bbox):
    x,y,w,h = int(bbox[0]),int(bbox[1]),int(bbox[2]),int(bbox[3])
    cv2.rectangle(image,(x,y),(x+w,y+h),(0,255,0),1)

#now start the loop
while camera.isOpened():
    #get the frame
    ret,frame = camera.read()
    frame = cv2.resize(frame,(500,500))
    
    #reference time count
    timer = cv2.getTickCount()
    
    
    #now get the box coordinates
    success,bbox = tracker.update(frame)
    
    #calculate the fps
    fps = cv2.getTickFrequency()/(cv2.getTickCount()-timer)
    
    #write the frame per second value on the frame
    cv2.putText(frame,str(int(fps)),(70,70),cv2.FONT_HERSHEY_COMPLEX,1,(0,255,0),1)
      
    if success:
        drawBox(frame,bbox)
    else:
        cv2.putText(frame,"lost",(70,80),cv2.FONT_HERSHEY_COMPLEX,1,(0,255,0),1)
    cv2.imshow("frame",frame)
    if cv2.waitKey(100) & 0xFF == 27:
        break
        
camera.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

cap = cv2.VideoCapture("new.avi")
ret,frame = cap.read()
frame = cv2.resize(frame,(500,500))
box = cv2.selectROI(frame,False)
tracker = cv2.TrackerMIL_create()

tracker.init(frame,box)

def drawbox(img,box):
    x,y,w,h = int(box[0]),int(box[1]),int(box[2]),int(box[3])
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),1)

while cap.isOpened():
    ret,frame = cap.read()
    frame = cv2.resize(frame,(500,500))
    
    timer = cv2.getTickCount()
    
    success,box = tracker.update(frame)
    print(success)
    if success:
        drawbox(frame,box)
    else:
        cv2.putText(frame,"lost",(10,40),cv2.FONT_HERSHEY_PLAIN,1,(0,255,0))
    
    fps = cv2.getTickFrequency()/(cv2.getTickCount()-timer)
    
    cv2.putText(frame,"fps:"+str(int(fps)),(10,10),cv2.FONT_HERSHEY_PLAIN,1,(0,255,0))
    
    cv2.imshow("video",frame)
    if cv2.waitKey(100) & 0xFF == 27:
        break
cap.release()
cv2.destroyAllWindows()
```
